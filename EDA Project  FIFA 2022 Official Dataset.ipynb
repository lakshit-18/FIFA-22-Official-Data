Here we can see that this dataset contains some columns that are not necessarily required for analysis some of those are
- Photo
- Flag 
- Club logo

So we would just delete these columns

```python
#Dropping the 'Photo', 'Flag' and 'Club Logo' columns
df.drop(['Photo', 'Flag', 'Club Logo'], axis = 1, inplace = True)
```

The 'ID' column also doesn't add any particular value to our dataset, we can just drop this column

```python
# Dropping the 'ID' column
df.drop(['ID'], axis = 1, inplace = True)
```

```python
df.Position = df.Position.apply(lambda x: x.split('>')[1])
df.head()
```

# Dealing with missing values

```python
df.isnull().sum()
```

### 1) Club column

```python
#Finding the number of null values in the 'Club' column
df.Club.isnull().sum()
```

```python
# Shape of the dataset
df.shape
```

```python
# Now we will find the percentage of the missing values in the 'Club' variable
float(100.0 * 264/16710)
```

So here we can see that the percenage is still very small and there is no other useful way to just impute the missing values
So just dropping the rows with the club missing would be the best option.

#### Dropping the records with Club missing

```python
df = df[~df.Club.isnull()].copy()
```

```python
df.isnull().sum()
```

### 2) Body Type

```python
df.Body Type.isnull().sum()
```

```python
# Renaming the column 'Body Type'
df = df.rename(columns={'Body Type': 'BodyType', 'Weak Foot': 'WeakFoot'})
```
